```python
import json

with open("results.json") as f:
    res = json.load(f)
```

```python
from sklearn.metrics import roc_curve

key = "DDM|{'drift_threshold': 0.1, 'warm_start': 10}"
fpr, tpr, thresh = roc_curve(res[key]["y_true"], res[key]["y_pred"])
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 5, figsize=(20, 4))

models = [
    "DDM",
    "EDDM",
    "FHDDM",
    "HDDM_A",
    "HDDM_W",
]

for key in res:
    pos = models.index(key.split("|")[0])
    fpr, tpr, thresh = roc_curve(res[key]["y_true"], res[key]["y_pred"])
    ax[pos].plot(fpr, tpr)

# plt.plot(fpr, tpr)
```
